```python
y = []
for index in json:
    value = index[2]
    if value == "true":
        y.append(1)
    else:
        y.append(0)
    
    
df = pd.read_json("ShortendTextRecData.json")
x = df.drop(axis=1,labels="isShortHand")
x = x.to_numpy()
y = np.array(y)
print("yay")
xw = np.array([np.random.random(161)])

def char_to_int(char):
    if char == 'a':
        return 1
    elif char == 'b':
        return 2
    elif char == 'c':
        return 3
    elif char == 'd':
        return 4
    elif char == 'e':
        return 5
    elif char == 'f':
        return 6
    elif char == "g":
        return 7
    elif char == 'h':
        return 8
    elif char == 'i':
        return 9
    elif char == 'j':
        return 10
    elif char == 'k':
        return 11
    elif char == 'l':
        return 12
    elif char == 'm':
        return 13
    elif char == 'n':
        return 14
    elif char == 'o':
        return 15
    elif char == 'p':
        return 16
    elif char == 'q':
        return 17
    elif char == 'r':
        return 18
    elif char == 's':
        return 19
    elif char == 't':
        return 20
    elif char == 'u':
        return 21
    elif char == 'v':
        return 22
    elif char == 'w':
        return 23
    elif char == 's':
        return 24
    elif char == 'x':
        return 25
    elif char == 'y':
        return 26
    elif char == 'z':
        return 27
    elif char == ' ':
        return 28
    
    
def str_to_ints(strng):
    result = []
    res = [] 
    res[:] = strng
    for char in res:
        result.append(char_to_int(char))
    return result
    
    
def convert(list): 
      
    # Converting integer list to string list 
    s = [str(i) for i in list] 
      
    # Join list items using join() 
    res = int("".join(s)) 
      
    return((res))


for i in x:
    a = str_to_ints(i[0])
    i[0] = convert(a)
    
    b = str_to_ints(i[1])
    i[1] = convert(b)
    
    
    
x = np.array(x, dtype=float)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
classifier = MLPClassifier(max_iter=1000)
classifier.fit(x_train,y_train)
pred = classifier.predict(x_test)
correct = 0
total = pred.size
for i in range(pred.size - 1):
    model_pre = pred[i]
    actrual = y_test[i]
    if model_pre == actrual:
        correct += 1
        
print(correct, total)
print("testing Acc: {}".format(correct / total))
```

```python
from sklearn.model_selection import GridSearchCV

def integers(a, b):
    return list(range(a, b+1))
    
def make_model(optimizer="Adam",layers=2,n_neurons=70,activation="relu"):
    model = keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=x_train.shape[1:])) 
#     model.add(keras.layers.Input(shape=x_train.shape[1:]))
    for i in range(layers):
        model.add(keras.layers.Dense(n_neurons, activation=activation))   
        
    model.add(keras.layers.Dense(2, activation="softmax"))
    model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer=optimizer, metrics=["accuracy"])
    return model


params = {
    "optimizer": ['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam'],
    "layers": integers(0,6),
    "n_neurons": integers(30,400),
    "activation": ["relu", "selu", "tanh","elu"]
}
seq = keras.wrappers.scikit_learn.KerasClassifier(make_model())
grid = GridSearchCV(estimator=seq, param_grid=params, n_jobs=-1, cv=3)
# grid_result = grid.fit(x_train, y_train)
```
